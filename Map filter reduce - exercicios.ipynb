# Testar as funções map, filter e reduce

```python
import numpy as np
```

```python
dados = np.random.randint(1,1000,100); dados
```

## Map - aplica uma função a um conjunto de elementos
### Criar uma função que eleve a lista ao quadrado com um lambda e map

```python
list(map(lambda x: x**2, dados))
```

### Criar uma função que retorne o código asci da string digitada.

```python
def cod(x):
    return ord(x)

palavra = input('Digite uma palavra para saber seu código asci\n')
l = [palavra[x] for x in range(len(palavra))]

list(map(cod, l))
```

## Filter - Similar a map, mas retorna apenas os valores que estejam de acordo com a condição
### Criar uma função que retorne se um número é par ou não

```python
def e_par(x):
    return x % 2 == 0

list(filter(e_par, dados))
```

## Reduce 
### Aplica uma função a uma lista de elementos, reduzindo - os até que reste apenas um elemento

```python
from functools import reduce 

def soma(x,y):
    return x+y

reduce(soma, dados)
```
